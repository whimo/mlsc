```python
import numpy as np
from skimage.io import imread
import skimage
```

```python
image = imread('parrots.jpg')
```

```python
import pylab
%matplotlib inline
```

```python
pylab.imshow(image)
```

```python
image = skimage.img_as_float(image) 
orig_image = np.copy(image)
shape = image.shape
```

```python
X = np.reshape(image, (shape[0]*shape[1], 3))
```

```python
from sklearn.cluster import KMeans

for n in range(1, 20):
    cls = KMeans(n, init='k-means++', random_state=241)
    pred = cls.fit_predict(X)

    reduced_image = np.empty((shape[0]*shape[1], 3))
    for i, cluster in enumerate(pred):
        reduced_image[i] = cls.cluster_centers_[cluster]
        
    reduced_image = np.reshape(reduced_image, shape)
    mse = np.mean((orig_image - reduced_image) ** 2)
    psnr = 10 * np.log10(1.0 / mse)
    
    print n, psnr
```
